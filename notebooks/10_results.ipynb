### Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
```

### Constants

```python
plotdir = Path("/Users/tusharsingh/Work/Project/tcga-mldl/results/plots")
```

### Results Data

```python
data = {
    'Split': ['5pct_80', '15pct_80', '25pct_80'] * 4,
    'Model': ['Logistic', 'Logistic', 'Logistic',
              'SVM', 'SVM', 'SVM',
              'Random Forest', 'Random Forest', 'Random Forest',
              'CNN', 'CNN', 'CNN'],
    'Accuracy': [0.6494, 0.7671, 0.7896,   # Logistic
                 0.6742, 0.7038, 0.7038,   # SVM
                 0.6595, 0.6620, 0.6376,   # RF
                 0.70, 0.73, 0.79]         # CNN
}
df_results = pd.DataFrame(data)
```

### Plot

```python
plt.figure(figsize=(8,6))
colors = ["#55efc4","#81ecec","#74b9ff","#a29bfe"]
sns.barplot(data=df_results, x='Split', y='Accuracy', hue='Model', palette=colors , edgecolor="black")
plt.ylim(0.60, 0.85)
plt.title('Model Accuracy Comparison Across Data Splits', fontsize=16, pad = 10)
plt.ylabel('CV Accuracy', fontsize=12, labelpad=10)
plt.xlabel('Data Split', fontsize=12 , labelpad=10)
plt.legend(title='Model', bbox_to_anchor=(0.01, 1) , loc='upper left')
plt.tight_layout()
plt.savefig(plotdir/"Acc_comparison-v-all.png", bbox_inches="tight", dpi=300)
plt.show()
```
